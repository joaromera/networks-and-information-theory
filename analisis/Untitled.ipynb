```python
from pathlib import Path
import pandas as pd
```

```python
redes = ['flinari', 'jjunqueras', 'jcperez', 'jromera']
experimentos = ['tarde', 'noche']

def read(red: str, experimento: str):
    columns = ['dest', 'protocolo', 'probabilidad', 'informacion']
    path = Path('datos') / experimento / f'{red}.csv'
    return pd.read_csv(path, names=columns)

def porcentaje_unicast_broadcast(df):
    return df.groupby('dest').sum()['probabilidad']
```

```python
for red in redes:
    for exp in experimentos:
        
        df = read(red, exp)
        print(red, exp)
        print(porcentaje_unicast_broadcast(df))
        
```
